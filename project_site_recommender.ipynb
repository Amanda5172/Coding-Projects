```python
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

sw = set(stopwords.words('english'))

def black_txt(token):
    return  token not in sw and token not in list(string.punctuation) and len(token)>2 

def clean_txt(text):
    clean_text = []
    clean_text2 = []
    text = re.sub("'",'',text) 
    text = re.sub("(\\d|\\W)+"," ", text) 
#     clean = re.compile('<.*?>')
#     text = re.sub(clean,'',text)
    clean_text = [nltk.stem.WordNetLemmatizer().lemmatize(word,pos="v") for word in word_tokenize(text.lower()) if black_txt(word)]
    clean_text2 = [word for word in clean_text if black_txt(word)] 
    return " ".join(clean_text2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
from sklearn.metrics.pairwise import cosine_similarity 
tfidf_all = tfidf_vectorizer.fit_transform(X['text'])
```

```python
def score(title):
    global output
    
    print(title['text'])
    user_tfidf = tfidf_vectorizer.transform(title['text'])
    cos_similarity_tfidf = map(lambda x: cosine_similarity(user_tfidf, x),tfidf_all) 
    output = list(cos_similarity_tfidf)
    
```

```python
def recommend(top, dfnew, scores):
    recommendation = pd.DataFrame(columns = ['title', 'score']) 
    count = 0
    for i in top:
        recommendation.at[count, 'title'] = dfnew['title'][i] 
        recommendation.at[count, 'score'] =  scores[count]
        count += 1
    return recommendation
```

```python
def get_recommendation(title,key):
    global output 
    df = pd.DataFrame(columns=[key])
    df[key] = [title]
    score(df)
    top = sorted(range(len(output)), key=lambda i: output[i], reverse=True)[:11] 
    
    list_scores = [output[i][0][0] for i in top]
    rec=recommend(top,dfnew, list_scores)
    
    return rec
```
